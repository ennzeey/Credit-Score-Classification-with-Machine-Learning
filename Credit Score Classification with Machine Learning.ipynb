## Credit Score Classification Model

```python
# Lets do some feature engineering by converting "Credit Mix" column into numerical so it can be used in our ml model"

#mix = {'Standard':1,'Good':2,'Bad':0}

df['Credit_Mix'] = df['Credit_Mix'].map({"Standard":1,"Good":2,"Bad":0})

from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
df.columns
```

```python
df.Age.max()
```

```python
bins = [0,18,25,35,45,55,float('inf')]
labels = ['0-18','19-25','26-35','36-45','46-55','56+']
df['Age_Range'] = pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
sns.countplot(x='Age_Range', data=df, hue= "Credit_Score", palette={'Poor': 'red','Standard':'yellow','Good':'green'})

plt.figure(figsize=(17,8))
plt.title('Age distribution by Credit Score',fontsize=17)
plt.xlabel('Age Range',fontsize=12)
plt.ylabel('volume')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

Ages between 26-35 got the most people with standard score as well as poor credit score, Ages between 36-45 also as a Good amount of Standard score and a lower poor socre as compared to 26-35

```python
plt.figure(figsize=(17,9))
sns.boxplot(x='Age_Range',y='Monthly_Inhand_Salary',data=df)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('Age Range',fontsize=13)
plt.ylabel('Monthly_Inhand_Salary',fontsize=13)
```

Ages between 46-55 has more people with Higer monthly inhand salary 

```python
df[df['Age_Range'] == '0-18']
```

```python
df['Age'] = df['Age'].round().astype(int)
```

```python
df['SSN'] = df['SSN'].round().astype(int)
```

```python
X = df[['Annual_Income','Monthly_Inhand_Salary','Num_Bank_Accounts',
       'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan',
       'Delay_from_due_date', 'Num_of_Delayed_Payment', 
       'Outstanding_Debt','Credit_History_Age', 'Monthly_Balance','Total_EMI_per_month','Amount_invested_monthly',"Age"
                 ]]

y= df[['Credit_Score']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X.head()
```

# Using Logistic Regression Classifier 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_pred = lr_model.predict(X_test)
```

```python
print(classification_report(y_test,lr_pred))
```

# Using RandomForest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize the model
rfc_model = RandomForestClassifier()
```

```python
rfc_model.fit(X_train,y_train)
```

```python
rfc_pred = rfc_model.predict(X_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

# Using DecisionTree Classifier 

```python
from sklearn.tree import DecisionTreeClassifier
# Initialize the model
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
dtree_pred = dtree.predict(X_test)
```

```python
print(classification_report(y_test,dtree_pred))
```
